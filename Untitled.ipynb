```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as optim

xk=[5,7,15,17,25,30,32,42,45,50]
#print(xk)
yk=[0.2,1.8,6.5,9.1,10.8,12.2,11.4,12.6,12.8,13]
xi=np.linspace(5,50)
def f(x, a, b, c):
    return c / (1 + a * np.exp(-b*x))

(a,b,c),cov = optim.curve_fit(f, xk, yk, np.random.exponential(size=3))
#print(c)
#y = f(xi,p1[1],p1[0])
y = f(xi,a,b,c)
plt.plot(xk,yk,'o')
plt.plot(xi,y, label = 'logistic model')
plt.legend(loc="upper right")

plt.show()
```
